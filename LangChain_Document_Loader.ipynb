```python
import yt_dlp

url = "https://www.youtube.com/watch?v=vStJoetOxJg&list=PLkDaE6sCZn6FNC6YRfRQc_FbeQrF8BwGI&index=1"

# this will download the best m4a audio
ydl_opts = {
    "format": "bestaudio/best",
    "outtmpl": "lecture.m4a"
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])
```

```python
import torch
```

```python
import whisperx


model = whisperx.load_model(
    "base",
    device="cuda" if torch.cuda.is_available() else "cpu",
    compute_type="float32"
)

result = model.transcribe("lecture.m4a")

full_text = " ".join([seg["text"] for seg in result["segments"]])
print(full_text[:1000])  # print first 1000 characters
```
